### Importando API-KEY

```python
from google.colab import userdata
servienvio = userdata.get('API-KEY-SERVIENVIO')
```

 Importando Biblioteca ChatGoogleGenerativeAI

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(
    model = "gemini-3-flash-preview",
    temperature = 0,
    google_api_key = servienvio
)
```

```python
respuesta = llm.invoke("que es la ia")
```
